```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn
```

```python
# This script demonstrates various Matplotlib visualizations such as scatter plots,
# bar plots, line plots, and histograms. It also customizes titles, labels, fonts,
# markers, line styles, and grid lines.
```

```python
# Set a random seed for reproducibility
np.random.seed(613)
# Generate data for visualization
x = np.arange(50) # 0,1,2,3,4,...,49
y = np.random.randint(0,100,50) # Random integers between 0 and 99 (50 values)
```

```python
# Scatter plot: Visualizes the distribution of y values against x
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)  # Plot points as individual markers
```

```python
# Bar plot: Represents y values as vertical bars
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y) 
```

```python
# Line plot: Connects y values with lines to show trends over x
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Histogram: Shows the frequency distribution of y values
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Customizing titles and axis labels for the line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title("Total growth over time")  # Title for the plot
ax.set_ylabel("Total growth")  # Label for the Y-axis
ax.set_xlabel("Years since start")  # Label for the X-axis
```

```python
# Define font styles for titles and labels
font1 = {"family": "sans-serif","color":"blue", "size":20}
font2 = {"family": "monospace","color":"green", "size":14}

# Apply font styles to titles and labels
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title("Total growth over time", fontdict = font1)
ax.set_ylabel("Total growth", fontdict = font2)
ax.set_xlabel("Years since start", fontdict = font2)
```

```python
# Scatter plot with customized markers
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
          marker ="*", # "o"/ "+"  # Star-shaped marker
          color = "indigo")  # Marker color
```

```python
# Line plot with customized marker and line styles

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y,
        marker="*",  # Star-shaped marker
        color="indigo",  # Line color
        linestyle="--",  # Dashed line
        linewidth=2,  # Line thickness
        markeredgecolor="#FA9359",  # Marker edge color
        markerfacecolor="#000000")  # Marker fill color
```

```python
# Marker color
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", # line color
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", # marker edge color
           markerfacecolor = "#000000") # marker inside colo
```

```python
# Adding grid lines to the plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", 
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", 
           markerfacecolor = "#000000") 
ax.grid() # both vertical and horizontal grid lines
#ax.grid(axis = "x")  # Uncomment for vertical grid lines only
# ax.grid(axis="y")  # Uncomment for horizontal grid lines only
```

```python
import matplotlib.pyplot as plt

# Sample data
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro" , "Nami" , "Usopp" , "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

# Create subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# Customize the scatter plot (ax1)
ax1.scatter(x1, y1, color='blue', marker='o', label='Data Points')
ax1.set_title("Scatter Plot", fontsize=14, fontweight='bold')
ax1.set_xlabel("episode number", fontsize=12)
ax1.set_ylabel("minutes", fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)  # Add gridlines
ax1.legend(loc='upper left')  # Add a legend
ax1.annotate("Point A", xy=(2, 5), xytext=(2.5, 5.5),
             arrowprops=dict(arrowstyle="->", color='black'))

# Customize the bar chart (ax2)
ax2.bar(x2, y2, color=['green', 'orange', 'red'], alpha=0.7, edgecolor='black')
ax2.set_title("Bar Chart", fontsize=14, fontweight='bold')
ax2.set_xlabel("Categories", fontsize=12)
ax2.set_ylabel("Values", fontsize=12)
ax2.text(1, 8, "Category B", fontsize=10, color='darkred')  # Add a text annotation

# Adjust layout
plt.tight_layout()
plt.show()
```
