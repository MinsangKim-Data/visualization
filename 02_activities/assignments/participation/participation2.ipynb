```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x, y1, color = 'indigo')
ax.plot(x, y2, color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2, yerr = y2_sd, fmt = "none",
            ecolor = 'indigo', elinewidth = 4, capsize = 6 , capthick = 4
            , errorevery= 2)
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid')

sns.lineplot(data=tips, 
x='total_bill', 
y='tip') 
```

```python
tipgraph = sns.lineplot(data=tips,
x='total_bill', y='tip')

tipgraph.set(title='Tips vs. Total Bill'
,xlabel='Total Bill ($)', ylabel='Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth =3, 
                        marker = 'o',
                        markerfacecolor = 'indigo')
```

```python
tipgraph = sns.scatterplot(data=tips, 
                           x='total_bill',
                           y='tip',
                           style = 'time',
                           hue ='day', 
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
sns.pairplot(data = tips,hue = 'day')
```

```python
daysplot = sns.relplot(data=tips, #creates relational plot
x="total_bill", # uses total_bill as x variable
y="tip", # uses tip as y variable
hue="sex", # differentiates points based on sex by color
col="day", # creates separate subplots for each unique value in the day column.
kind="scatter", # the plot type is a scatterplot
col_wrap=2) # want in 2 columns
```

```python
import plotly.graph_objects as go # 'go' is 'graph objects’
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(
title="Pirate Scores", xaxis_title="Pirates", yaxis_title="Score")

graph.show()
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers'
, # we want points for a scatter plot
marker=dict(
size=15, # point size
color='hotpink'
, # point colour
opacity=1, # point transparency/alpha
line=dict(width=5, color='purple') # point outline
)))
graph.update_layout(
title='Interactive Pirate Plot'
,
xaxis_title='Pirates'
,
yaxis_title='Scores'
,
width=500, height=500)
```

```python
from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)
# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
plt.show()
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B],
set_labels = ('Fruits','Pies'),
set_colors=("blue","red"), alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))

plt.show()
```
